```python
#Question 7
# How many unique countries are there in western europe

columns = ["Regional indicator", "Country name"]
data = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/world-happiness-report-2021.csv", usecols=columns)
def main(data):
    new_list = []
    for k, i in zip(list(data["Regional indicator"]), range(len(data))):
        if k == 'Western Europe':
            if list(data['Country name'])[i] not in new_list:
                new_list.append(list(data['Country name'])[i])
    return new_list
unique_countries = main(data)
print(len(unique_countries))
```

```python
#topic 2
#Question 1
# Display the data of top 5 east asian countries based on 'Generosity'

columns = ["Regional indicator", "Country name", "Generosity"]
data = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/world-happiness-report-2021.csv", usecols=columns)
dict_data = {}
for i in range(len(data)):
    if 'East Asia' ==  list(data['Regional indicator'])[i]:
        dict_data[f"{ list(data['Country name'])[i] }"] = list(data['Generosity'])[i]

dict_data = {key: value for key, value in sorted(dict_data.items(), key=lambda item: item[1])}

num: int = 0
for k, v in reversed(dict_data.items()):
    if num == 5:
        break
    else:
        print(k,v)
        num += 1
```

```python
# topic2
# Question 2
# Build a plot (line plot) that shows the variation of 'Ladder score' among the above 5 east asian countries (based on generosity)

def data_main(sort: bool = True, region: str = 'South Asia'):
    columns = ["Regional indicator", "Country name",
               "Generosity", "Ladder score"]
    data = pd.read_csv(
        "https://raw.githubusercontent.com/dphi-official/Datasets/master/world-happiness-report-2021.csv", usecols=columns)

    plt.figure(figsize=(8, 5))

    lists_gen = [[data.Generosity[i], list(data['Ladder score'])[i], list(data['Country name'])[
        i]] for i in range(len(data.Generosity)) if list(data['Regional indicator'])[i] == region]
    if sort:
        lists_gen = sorted(lists_gen, key=lambda tup: tup[1], reverse=True)
    else:
        pass
    generosity_lists, ladder_score_list, country_name_list = [], [], []

    for a, b, c in lists_gen:
        generosity_lists.append(a)
        ladder_score_list.append(b)
        country_name_list.append(c)
    return generosity_lists, ladder_score_list, country_name_list


generosity_lists, ladder_score_list, country_name_list = data_main(region='East Asia')

plt.figure(figsize=(8, 5))
plt.plot(country_name_list[:5], ladder_score_list[:5], marker='o')
plt.title("Variation of 'Ladder score' (based on Generosity)")
plt.xlabel("Countries Name")
plt.ylabel("Ladder Score")
plt.show()
```

```python
#Topic 2 
#Question 3
# Build a plot that shows the variation of 'Ladder score' among the 5 south east asian countries

generosity_lists, ladder_score_list, country_name_list = data_main(sort=False)

plt.plot(country_name_list[:5], ladder_score_list[:5], 'ro--')
plt.title("Variation of 'Ladder score'")
plt.xlabel("Countries Name")
plt.ylabel("Ladder Score")
plt.show()

```

```python
#topic 2
#Question 4
# Create a dataframe object : df_2021 with the following countries: 'China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan'.


#install ordered-set using [pip install ordered-set]
from ordered_set import OrderedSet
class df_2021:
    countries_name_gdp_tup = []

    def __init__(self) -> None:
        countries = ['China', 'Nepal', 'Bangladesh',
                     'Pakistan', 'Myanmar', 'India', 'Afghanistan']

        columns = ["Country name",'Logged GDP per capita']

        data = pd.read_csv(
            "https://raw.githubusercontent.com/dphi-official/Datasets/master/world-happiness-report-2021.csv", usecols=columns)

        df_2021.countries_name_gdp_tup = OrderedSet([ (list(data['Country name'])[i] ,list(data['Logged GDP per capita'])[i]) for i in range(len(data)) if list(data['Country name'])[i] in countries ])

        self.data_tuples = list((k,v) for k,v in df_2021.countries_name_gdp_tup)

    def dataframe(self):
        dataframe_countries_and_gdp = pd.DataFrame(self.data_tuples, columns=['countries','gdp'])
        return dataframe_countries_and_gdp



dataframe  = df_2021().dataframe
print(dataframe) #object
```

```python
#topic 2
#question 2 2nd part 
# Now, build a scatter plot to show the relation these countries vs their 'Logged GDP per capita'


plt.figure(figsize=(8,5))
plt.scatter(dataframe.countries, dataframe.gdp)
plt.xlabel('Countries')
plt.ylabel('Gdp')
plt.title("Countries vs Gdp")
plt.show()
```
